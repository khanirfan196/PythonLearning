<h2>Python Decorators</h2>

```python

def sum(a, b):
    return a + b


print(sum(2, 2))
```

```python
def divs(a ,b):
    return a / b

print(divs(4, 2))
```

```python
# Decorator to check denominator is zero 

def div_decorator(func_as_arg):
    def div_wrapper(a, b):
        if b == 0:
            b = 1
        return func_as_arg(a, b)
    return div_wrapper

@div_decorator
def division(a, b):
    return a / b

print("Division is: " + str(division(4, 0)))
```

```python
def calc_deco(funct):
    def calc_wrapper(x, y):
        if y == 0:
            y = 1
        return funct(x, y)
    return calc_wrapper


def addition(x, y):
    return x + y

def subtract(x, y):
    return x - y

@calc_deco
def multiplication(x ,y):
    print("value of y: " + str(y))
    return x * y

@calc_deco
def divisions(x, y):
    print("value of y: " + str(y))
    return x /y


def calculate(func, x, y):
    return func(x, y)

print(calculate(multiplication, 5, 0))
```

<h2>Decorators with arguments</h2>

<ul>
    <li>Usually a decorator is of 2 layers of functions - inner and outer.</li>
    <li>To have an argument in decorator, one more layer of function need to be added.</li>
</ul>

```python

def deco_args(prefix):
    def sum_deco(func):
        def sum_wrapper(prefix, x, y):
            if x < 0:
                x = -1 * x
            if y < 0: 
                y = -1 * y
            print("In wrapper function", prefix, x, y)
            return func(prefix, x, y)
        return sum_wrapper
    return sum_deco



@deco_args("")
def sum(prefix, x, y):
    z = x + y
    return prefix +" x + y value is: " + str(z)

print(sum("ADD", -5, -5))
```
