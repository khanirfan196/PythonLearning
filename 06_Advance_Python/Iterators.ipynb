```python
# An iterator is an object which contains 'n' number of countable values. 
# Exampe list, tuples, dictionaries, sets 
# As the above collectiosn are iterator objects, we can make them iterable by using
# iter() and next() methods
```

```python
sentence = "this is a sentence"

print(sentence)
```

```python
sentence = iter("this is a sentence")

print(next(sentence))
print(next(sentence))
print(next(sentence))
print(next(sentence))
print(next(sentence))
print(next(sentence))

# So by using next() method, the object 'sentence' becomes
# iterable and it keeps printing until the last value is met. 
# This gives some control over iterable objects.
```

```python
fruits = ["apple", "orange", "banana", "kiwi", "pineapple"]

iter_fruits = iter(fruits)


# or with loop

while True:
    items = next(iter_fruits, None)
    if items is None:
        print("List completed")
        break
    else:
        print(items)
```

```python
# Iterator Class

class IterClass:
    def __init__(self, current, lastval):
        self.current = current
        self.lastval = lastval
        
    def __iter__(self, current):
        self.current = current
        
    def __next__(self):
        if self.current < self.lastval:
            self.current += 1
            return self.current
        else:
            raise StopIteration


itr = IterClass(1, 6)

while True:
    x = next(itr, None)
    if x == None:
        break
    else:
        print(x)
```
