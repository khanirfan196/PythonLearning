```python
# To read, create, update a file, python provides open() function

# open() take two arguments. One is the file path and other is the mode 
# like r, a, w, x.
# r - read
# w - write
# a - append
# x - creates a file

# the default mode is r if nothing specified.
```

```python
# Read from a file

f = open("sample_files/hello_world.txt")

f.read()

f.close()
```

```python
# Write to a file

# Writing can be done with mode append or write. 

# write mode - it overwrites the file content

f_w = open("sample_files/hello_world.txt", 'w')

sentence = "This line is written from code."

f_w.write(sentence)
```

```python
# append mode

f_a = open("sample_files/hello_world.txt", "a")

sentence_a = "\nThis line is appended from code.\n"

f_a.write(sentence_a)
```

```python
f_a.close()
```

```python
# 'with' block for file handling. It provides a decent way of reading 
# or writing to files. The open() just opens the files and reads it. 
# But it does not close it. Using 'with' block, all operations can be
# performed efficiently and neatly. 


with open("sample_files/demo.txt", 'r') as fr:
    print(fr.read())
    fr.close()
    
```

```python
# Along with read() method, there is another method called readline().
# This method reads only first line of the file. 

with open("sample_files/demo.txt", 'r') as fr:
    print(fr.readline())
    fr.close()
```

```python
# Appending data to file by using 'with' block

data_line = "This is appended data from code."

with open("sample_files/demo.txt", 'a') as fa:
    fa.write(data_line)
    fa.close()
```
