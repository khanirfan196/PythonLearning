```python
# Creating List with list constructor or copy() method

# List Constructor
```

```python
new_list = list("HelloWorld")

print(new_list)

# List Constructor converts a string into list with individual characters
```

```python
newlist2 = list(("car", "bike", "ship", "plane", "truck", "rocket"))

print(newlist2)

# Above is a tuple passed to list constructor
```

```python
# copy() method to copy from existing list and create new one

vehicle_list = newlist2.copy()

print(vehicle_list)
```

```python
# List Iteration and Indexing.

# This is same as a String Iteration

# First element
print(newlist2[0])

# Last element
print(newlist2[-1])

# Starts from 0 and ends at 4th index value
print(newlist2[:4])

# Reverse Indexing

# 4th value in reverse
print(newlist2[-4])

# Starts from 4th value in reverse and prints all the way except the last value
print(newlist2[-4:-1])
```

```python
# range() method to just get the list indexes

for i in range(len(newlist2)):
    print(i)
```

```python
# Changing List Values

newlist2[1] = "motorbike"

# 'bike' got replaced with 'motobrike' value
print(newlist2)
```

```python
# Changing values through ranges

newlist2[0:2] = ["sedan", "sportsbike", "voyager"]

# Frist 3 values got replaced
print(newlist2)
```

```python
# Inserting Values in list

# Insert() method

newlist2.insert(2, "titanic")

# 'titanic' got added at index 3 and all values got new indexes
print(newlist2)
```

```python
# Append() method

newlist2.append("spaceship")

# 'spaceship' value got appended at the end of list
print(newlist2)
```

```python
# Extend() method

newlist3 = ["suv", "crane", "train"]

newlist2.extend(newlist3)

# Appends the other list values to the end of the main list
print(newlist2)
```

```python
# We can use '+' operator to join to lists

list1 = ["hello"]

list2 = ["world"]

print(list1 + list2)
```

```python
# Removing Items from List

# remove() method

# removes specified item
newlist2.remove("voyager")

print(newlist2)
```

```python
# pop() method

newlist2.pop(0)

# removes item at specified index.
print(newlist2)
```

```python
newlist2.pop() 

# If no index is specified, removes the last item by default
print(newlist2)
```

```python
# clear() method

# makes the list empty
newlist3.clear()

print(newlist3)
```

```python
# 'del' keyword to delete the list

del newlist3

# Below print statement throws error for not finding the above list
print(newlist3)
```

```python
# List Comprehensions - Shortest way to generate list based on conditions or methods

# Generating a list with five integer elements

int_list = [x for x in range(1, 6)]

print(int_list)
```

```python
# Generating a list based on other list and certain condition

cars_list = ["toyota_camry", "toyota_corolla", "kia_optima", "honda_fit"]

toyota_cars = [cars for cars in cars_list if "toyota" in cars]

print(toyota_cars)
```

```python
# Sorting List

# sort() method

list_alps = ['a', 'c', 'e', 'f', 'b', 'd']

print(list_alps)

list_alps.sort()

print(list_alps)

# sort() method with reverse keyword

list_alps.sort(reverse=True)

print(list_alps)
```

```python

# sort() method can be used by passing a keyword function 

fruits_list = ["banana", "Apple", "kiwi", "Mango", "orange"]

fruits_list.sort(key = str.lower)

print(fruits_list)
```

```python

# Reversing a list - This reversing the order of elements in the list

fruits_list.reverse()

print(fruits_list)
```
