```python
# Creating tuple with tuple() constructor

tuple2 = tuple(("charis", "desk", "lamp", "books"))

print(tuple2)
```

```python
# Accessing Tuples

# Tuples can be accessed the same way as list, through positive and negative indexing.

# They can be accessed through indexing in either forward or backward direction

# Iterable and can be done using range() and len() functions through loops

# Tuples are immutable. Any element cannot be added or changed or removed

# There is no tuple comprehension like list
```

```python
# Unpacking Tuples

(item1, item2, item3, item4) = tuple2

print(item1)

print(item2)

print(item3)

print(item4)
```

```python
# Using asterisk makes the element to access rest of the all elements in the form of list

(item1, *item2) = tuple2 

print(item1)

print(item2)
```

```python
# Assigns last element of tuple to last item and then gives all the other elements to item2

(item1, *item2, item3) = tuple2 

print(item1)

print(item2)

print(item3)
```

```python
# Joining Tuple

tuple_fruits = ("apple", "mango", "banana")

tuple_veggies = ("okra", "tomato", "potato")

print(tuple_fruits + tuple_veggies)
```

```python
# Counting specified element in a tuple for number of occurences

tuple_nums = (1, 2, 3, 4, 5, 6, 6, 6, 7)

# Number 6 appeared 3 times
print(tuple_nums.count(6))
```

```python
# index() methods returns the index of first occurence of element

# The first index of number 6 is 5. 
print(tuple_nums.index(6))
```
