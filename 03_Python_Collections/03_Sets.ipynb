```python
# Sets are another built-in datatype

# They are unchangeable and unordered. Doesn't allow duplicate values unlike list and tuple.

# But elements can be added and removed. They cannot be changed. 
```

```python
thisset = {"pen", "book", "scale", "pencil", "pencil"}

# There is a duplicate value, but set internally treated it as single value
print(thisset)
```

```python
# set() constructor to create a new set

new_set = set(("shirt", "pant", "tshirt", "trousers"))

print(new_set)
```

```python
# Accessing Sets

for clothes in new_set:
    print(clothes)
```

```python
# Adding elements to set - add() method

new_set.add("socks")

print(new_set)
```

```python
# remove() method 

new_set.remove("tshirt")

print(new_set)
```

```python
# discard() method

new_set.discard("tshirt")

print(new_set)
```

```python
# difference between remove and discard method is that the 
# remove method throws an error if the element to be removed is 
# not present in set, where as discard() doesn't
```

```python
# pop() method 

# this method removes random element and no index can be passed

new_set.pop()

print(new_set)


new_set.pop()

print(new_set)
```

```python
# Set can be cleared and deleted like list and tuple

# clear() method

new_set.clear()

print(new_set)
```

```python
# del keyword

del new_set

# as set will be deleted, print statement will throw error
print(new_set)
```

```python
# Joining Sets

# update() method to join two sets or with other iterable objects like list, tuple, dictionaries

new_set = set(("shirt", "pant", "tshirt", "trousers"))

w_set = {"jackets", "boots", "gloves"}

new_set.update(w_set)

print(new_set)
```

```python
# union() method - this does the same as update() method.

# union() requires a new set variable.

w2_set = {"socks", "muffler", "inners"}

w3_set = new_set.union(w2_set)

# or w3_set = new_set | w2_set

print(w3_set)

# '\' (pipe) symbol can be used for joining sets as union
```

```python
# intersection() method

set1 = {1, 2, 3}

set2 = {2, 3, 4}

set3 = set1.intersection(set2)

print(set3)

# '&' symbol can be used for intersecting two sets

# set3 = set1 & set2
```
