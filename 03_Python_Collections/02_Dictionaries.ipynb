```python
'''
Docstring for 03_Python_Collections.02_Dictionaries.ipynb
A dictionary is a key, value data type.
'''
```

```python
my_dict = {
    "name" : "mike",
    "email": "mike@gmail.com",
    "ids" : ["passport", "drivers license"],
    "age" : 30
}
```

```python
print(my_dict)
```

```python
for key, value in my_dict.items():
    print(key, value)
```

```python
for keys in my_dict.keys():
    print(keys)
```

```python
for value in my_dict.values():
    print(value)
```
