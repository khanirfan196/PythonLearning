```python
# Exception Handling mechanism is to manage to dynamic errors in the code.
# It has following blocks.
# try
# except
# else
# finally

# Try block holds the main code where there could be potential error in run time.
# Except block holds the exception which gets raised when an error occurs.
# Else block executes after try block executes successfully.
# Finally block gets executed irrespective of whether an error or occurs or not in 'try' block.
```

```python
try:
    a = 5
    b = 0
    c = a / b
    print(c)
except ZeroDivisionError:
    print("Division by 0 is not possible")
```

```python
# with else block

def numdiv(a, b):
    try:
        c = a / b
        print(c)
    except ZeroDivisionError:
        print("Division by 0 is not possible")
    else:
        print("Division is successful")

numdiv(5,5)
```

```python
# with 'finally' block

def numdiv(a, b):
    try:
        c = a / b
        print(c)
    except ZeroDivisionError:
        print("Division by 0 is not possible")
    else:
        print("Division is successful")
    finally:
        print("Finally block executed")
        
numdiv(4, 2)
```

```python
# Explicitly raising exception using 'raise' keyword

def numcheck(x):
    if x < 0:
        raise ValueError("x cannot be less than 0")
    else:
        print("x is greater than 0")

numcheck(-1)
        
```
