```python
# Finding . or - specifically by using square brackets 

# square brackets is like range


pattern = re.compile(r'\d\d\d[.-]\d\d\d[.-]\d\d\d\d')

with open('data.txt', 'r') as f:
    contents = f.read()
    matches = pattern.finditer(contents)

    for match in matches:
        print(match)  
```

```python
# matching only numbers which starts with 800 or 900

# \d looks for 0-9 digits

pattern = re.compile(r'[89]00.\d\d\d.\d\d\d\d')

with open('data.txt', 'r') as f:
    contents = f.read()
    matches = pattern.finditer(contents)

    for match in matches:
        print(match)  
```

```python
multi_line = """
cat
mat
pat
bat
"""

pattern = re.compile(r'[a-z][a-z][a-z]')

matches = pattern.finditer(multi_line)

for match in matches:
    print(match)
```

```python
# patter to find using negate ^

# matches all words except which starts with b

pattern = re.compile(r'[^b]at')

matches = pattern.finditer(multi_line)

for match in matches:
    print(match)
```

```python
# Quantifiers 


pattern = re.compile(r'\d{3}.\d{3}.\d{4}')

with open('data.txt', 'r') as f:
    contents = f.read()
    matches = pattern.finditer(contents)

    for match in matches:
        print(match)  
```

```python
# match() function

sentence = 'Start a sentence and then bring it to an end'

pattern = re.compile(r'Start')

matches = pattern.match(sentence)

# print(matches.group())

# print(matches.span())

print(matches.string)
```

```python
# sub() replaces any sub string with provided value

# "b" gets replaced with "a"

sub_str = re.sub("a", "b", sentence)

print(sub_str)
```

```python
# findall()

# It returns a list of all elements matches in the string or search pattern

pattern = re.compile(r'e')

matches = pattern.findall(sentence)

print(matches)

print(type(matches))
```
