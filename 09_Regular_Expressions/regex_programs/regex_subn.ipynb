```python
import re
```

```python
# subn() works exactly as sub() but it also returns 
# the number of substitutions it made along with a new string
```

```python
text = "The rain in Spain falls mainly in the rainy season"
result = re.subn(r"rain", "snow", text)

print(type(result))

# it returns a tuple, which will have new string along with the
# number of counts it made in the string.
# sub() only returns the string
print(result)
```

```python
# example with count argument

text = "apple apple apple apple"
result = re.subn(r"apple", "orange", text, count=2)


# though word 'apple' is 4 times, but it replaced only 2 times
print(result)
```
