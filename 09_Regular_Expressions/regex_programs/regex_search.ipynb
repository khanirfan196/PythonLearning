```python
import re
```

```python

my_str = f"llo, Hello, Hello, World"

pattern = 'He'

my_search = re.search(pattern, my_str)

# search() returns only the first match 
```

```python

print(my_search)

# my_search is an object of class 'Match' 

print(type(my_search))
```

```python
my_str = f"Hello, Hello, Hello, World"

pattern = 'He'

my_search = re.findall(pattern, my_str)

# findall() returns all the matches as a list.
```

```python
print(type(my_search))
```

```python
print(my_search)
```
