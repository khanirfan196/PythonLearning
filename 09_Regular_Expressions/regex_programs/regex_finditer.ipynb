```python
import re
```

```python
my_string = """
123 these are three digits.
287 these are too three digits.
899 these are digits too.
 """

pattern = r"\d{3}"

my_match = re.finditer(pattern, my_string)

print(my_match)
```

```python
# Iterating 'my_match' which is an object.

for matches in my_match:
    # group() is nothing but the pattern itself, but it all the found patterns are 
    # under a group(). like 123, 287, 899 are groups. 
    # groups() and groupdict() does not work because regex needs to be under paranthesis.
    # refer programs or .py files with their names itself.
    print(matches.group(), matches.start(), matches.end(), matches.groups(), matches.groupdict())
```
