```python
import re
```

```python
# example 1

my_str = "Irfan is   learning  regex"

split_pattern = r"\s+"
# here the split function is using regex expression.
# notice the string has multiple spaces but still each character is returned.

my_chars = re.split(split_pattern, my_str)
```

```python
print(my_chars)
```

```python
# example 2

my_str = "apple,banana;mango;pears,orange"
# here string is separated with either ',' or ';'.

my_chars = re.split(r"[;,]", my_str)
```

```python
print(my_chars)
```
