```python
import re
```

```python
# escape() function escapes special characters and treats them as part of string.
# Generally in a string variable, we use '\\' double backslashes to escape a special
# character, the same thing is being done by escape()
```

```python
text = ".+()[]{}"
escaped = re.escape(text)

print("Original :", text)
print("Escaped  :", escaped)
```

```python
pattern = re.escape("()+-")

mystr = "()+-hello"

search = re.search(pattern, mystr)

print("Found special characters:", search.group())
```
