```python
import re
```

```python
# sub() is use to replace/substitute a word or char with regex or any string
# syntax: re.sub(pattern, replacement, string, count=0)
```

```python
my_str = "irfan hello khan hello is learning hello python"
pattern = r"hello"
replacement = ""

new_str = re.sub(pattern, replacement, my_str)

# 'replacement' can be a function as well
```

```python
print(new_str)
```

```python
# example of 'count' parameter 

my_str = "1234"
pattern = r"\d"
replacement = "#"

new_str = re.sub(pattern, replacement, my_str, count=2)

print(new_str)
```
