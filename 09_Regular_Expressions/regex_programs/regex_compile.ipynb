```python
import re
```

```python
my_str = '''
Hello this is python learning.
Hello this is python learning.
Hello this is python learning.
Hello this is python learning.
Hello this is python learning.
'''
```

```python
# compile() is just use to compile the regex expression and this expresssion
# can be use anywhere and multiple times.
# Just like a variable, defined and assigned a value once, then re-use it many times.

pattern = re.compile("python")
```

```python
my_search = re.search(pattern, my_str)
my_findall = re.findall(pattern, my_str)
my_match = re.match(pattern, my_str)
my_fullmatch = re.fullmatch(pattern, my_str)
my_finditer = re.finditer(pattern, my_str)
```

```python
print(my_search.group())
print(my_findall)
print(my_match)
print(my_fullmatch)
print(my_finditer)

for itr in my_finditer:
    print(itr.start(), itr.group(), itr.end())
```
