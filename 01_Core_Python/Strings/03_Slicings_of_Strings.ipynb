```python
# Strings can be access through indexing just like arrays
```

```python
a = "python is a wonderful programming language"

# character at index 0
print(a[0])
```

```python
# prints characters starting from index 0 till 9.

print(a[0:10])
```

```python
# Reverse access of string characters

# this prints last character in the string
print(a[-1])
```

```python
# prints everything except last character

print(a[0:-1])
```

```python
# prints entire string with negative starting index

print(a[-42:])
```

```python
#
#  0  1  2  3  4   -- positive index
#  h  e  l  l  o   -- String
# -5 -4 -3 -2 -1   -- negative index
#
```

```python
letters = "abcdefghijklmnopqrstuvwxyz"
```

```python
reverse_str = letters[::-1]
print(reverse_str)
```
