```python
"""
Python has the following data types built-in by default, in these categories:

Text Type:	str
Numeric Types:	 int, float, complex
Sequence Types:	 list, tuple, range
Mapping Type:	 dict
Set Types:	 set, frozenset
Boolean Type:	 bool
Binary Types:	 bytes, bytearray, memoryview
None Type:	 NoneType

"""
```

```python

# type() - To get the variable's datatype 

x = 10

print(x)

print(type(x))
```

```python
""" 
As python is dynamically typed language, we still explicitly provide data type to the 
variable by using the constructor functions. 

x = str("Hello World")	str	
x = int(20)	int	
x = float(20.5)	float	
x = complex(1j)	complex	
x = list(("apple", "banana", "cherry"))	list	
x = tuple(("apple", "banana", "cherry"))	tuple	
x = range(6)	range	
x = dict(name="John", age=36)	dict	
x = set(("apple", "banana", "cherry"))	set	
x = frozenset(("apple", "banana", "cherry"))	frozenset	
x = bool(5)	bool	
x = bytes(5)	bytes	
x = bytearray(5)	bytearray	
x = memoryview(bytes(5))	memoryview

"""
```

```python
x = str("Hello Python")

print(x)

print(type(x))
```

```python
"""
Also, above functions can be use to cast a variable from one data type to other.

"""
```

```python
x = "10"

print(type(x))

y = int(x)

print(type(y))
```
