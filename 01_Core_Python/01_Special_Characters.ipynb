```python
"""
Special Characters in Python 

. , [ ] { } ( ) - _ + = * & ^ % $ # @ ! ~ ' ' \t=tab space, \n=newline

"""
```

```python
print("hello")
```

```python

# \t provides tab space before printing characters

print("\thello")
```

```python

# \n provides new line

print("hello\nWorld")
```

```python

# 'end' parameter adds the provided value. It can be space, newline, characters or strings etc.

print("hello World ", end="ppp")
```

```python

# By default 'end' parameter has '\n' next line. If space is provided, the second line prints in the first line.

print("hello world", end=" ")
print("second line")
print("third line")
```
