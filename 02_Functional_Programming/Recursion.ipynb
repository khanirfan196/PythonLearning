```python
# Recursive Function

# A Function which calls itself is called a recursive function. 
```

```python
def recur_func(n):
    if n == 1:
        return n
    else:
        return n * recur_func(n - 1)
```

```python
# For any number, the calculation goes like this
# n = 5, --> 5 * 4 * 3 * 2 * 1
# Function stops until the value reaches 1
# So in case of value 5, the function has called itself 4 times. 

recur_func(5)
```
