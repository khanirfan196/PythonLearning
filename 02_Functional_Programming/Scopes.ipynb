```python
# Scope of a variable 

x = 200

def myfunc():
    x = 100
    print(x)
    
print(x)

myfunc()
```

```python
# Global Keyword

# It is use to declare any variabe global inside a scoped function.
# In the below example, variable y is declared as global and can be used
# outisde of function once it is called or variable is initiated.

def func2():
    global y
    y = 500

func2()

print(y)

```

```python
# nonlocal keyword

# It is use to declare a variable residing in an inner function as nonlocal to outside function.
# Upon calling the outer function, variable s got value 'Sammy' which is inside inner function.
# the nonlocal keyword makes the variable available to outside function. 

def outfunc():
    s = "Jamie"
    def innfunc():
        nonlocal s 
        s = "Sammy"
        return s
    innfunc()
    return s

print(outfunc())
```
