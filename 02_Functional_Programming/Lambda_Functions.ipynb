```python
# A lambda function is a small anonymous function.
# A regular function is a definition of statements 
# whereas a lambda function is just an expression. 
# As lambda functions are expressions, they are generally
# suited for small operations, but bad for complex logic.
```

```python
x = lambda a: a * a 
```

```python
print(x(5))
```

```python
y = lambda a, b: a + b
```

```python
print(y(2, 2))
```

```python
def lambdademo(n):
    return lambda a: a * n
```

```python
calledfunc = lambdademo(5)

# value 5 belongs to a
```

```python
# argument belongs to n

print(calledfunc(2))
```
