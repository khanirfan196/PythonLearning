```python
# Arbitrary Arguments, *args while calling a function
```

```python
def myfunc(*args):
    return "This is one of the arguments : " + args[1]
```

```python
print(myfunc("irfan", "sidra", "rayyan"))
```

```python
# Keywords Argumengs, which are in key, value pair.
```

```python
def keyfunc(key1, key2, key3):
    return "The first keyword argument is : " + key1
```

```python
print(keyfunc("car", "truck", "plane"))
```

```python
# Multiple/Arbitrary Keyword arguments, **kwargs
```

```python
def arbkeyfunc(**cars):
    return "The car name is : " + cars["car_name"]
    
```

```python
print(arbkeyfunc(car_name="camry", manufacturer="toyota", country="japan"))
```

```python
# We can pass list as an argument

fruits = ["mango", "apple", "banana"]

def list_read(param_list):
    for vals in param_list:
        print(vals)
```

```python
list_read(fruits)
```

```python
# Positional only arguments 

# To have a function which should only accept positional arguments,

# we can provide '/' after argument 
```

```python
def onlyposarg(x, y, /):
    print("First positional argument is : " + x)
    print("Second positional argument is : " + y)
```

```python
onlyposarg("hello", "world")
```

```python
# Keyword only arguments
```

```python
def keyonlyarg(*, x, y, z):
    print("First keyword argument is : " + x)
    print("Second keyword argument is : " + y)
    print("Third keyword argument is : " + z)
    
```

```python
# if asterisk '*' is removed from above function,
# we can pass arguments without specifiying its positional parameters.

# keyonlyarg("IT","CS","CIVIL")
```

```python
keyonlyarg(x="IT", y="CS", z="CIVIL")
```

```python
# Combinatin of positional and keyword only parameters 

# This can be achieived using both '/' and '*'
```

```python
def combofunc(a, b, /, *, keyargs):
    print("First positinal parameter : " + a)
    print("Second positinal parameter : " + b)
    print("First keyword parameter : " + keyargs["x"])
    print("Second keyword parameter : " + keyargs["y"])
```

```python
combofunc("hello", "world", y="CAR", x="TRAIN")
```
